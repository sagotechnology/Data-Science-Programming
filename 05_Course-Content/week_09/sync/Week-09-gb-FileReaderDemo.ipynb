<h1 style="color:red">There are 6 access modes in python.</h1>

<ol><li><code>Read Only (‘r’)</code> : Open text file for reading. The handle is positioned at the beginning of the file. If the file does not exists, raises I/O error. This is also the default mode in which file is opened.
    </li><li><code>Read and Write (‘r+’)</code> : Open the file for reading and writing. The handle is positioned at the beginning of the file. Raises I/O error if the file does not exists.
</li><li><code>Write Only (‘w’)</code> : Open the file for writing. For existing file, the data is truncated and over-written. The handle is positioned at the beginning of the file. Creates the file if the file does not exists.
</li><li><code>Write and Read (‘w+’)</code> : Open the file for reading and writing. For existing file, data is truncated and over-written. The handle is positioned at the beginning of the file.
</li><li><code>Append Only (‘a’)</code> : Open the file for writing. The file is created if it does not exist. The handle is positioned at the end of the file. The data being written will be inserted at the end, after the existing data.
    </li><li><code>Append and Read (‘a+’)</code> : Open the file for reading and writing. The file is created if it does not exist. The handle is positioned at the end of the file. The data being written will be inserted at the end, after the existing data.
    </li></ol>

```python
''' a function to read records from a file using a generator and yield. 
note that yield can be used when calculating/processing large chunks of data '''

def readMyData(inputFileName, delimiter = ","):
    with open(inputFileName, 'r+b') as f:
        ''' below: raverse sequentially through the file '''
        for record in f:
            ''' the parsing logic for files (binary, text, json, xml, etc.) '''
            x = record.split(delimiter)
            ''' get the stream of things
             (e.g., words in the line of a text file,
             or fields in the row of a CSV file) '''
            yield x  # yield suspends a function and returns a value
                     # but stays in this state ... like. adebugger
            
''' another generator '''
def fibonacci(n):
    curr = 1
    prev = 0
    counter = 0
    while counter < n:
        yield curr
        prev, curr = curr, prev + curr
        counter += 1

print(fibonacci(5))
```

<h1 style="color:'red';">Reading and timing really large files</h1>
<a href="https://github.com/taupirho/read-big-file-with-python/blob/master/read_big_file.py" target="new">From GitHub</a>

```python
'''
The code below was run in a Jupyter notebook on an i7 quad core windows PC 
with 16GB of RAM. The input file was 24 Gbytes in size with approx 335 Million records
'''

from itertools import islice
from collections import defaultdict


def get_chunk_of_lines(file, N):
    """
    Retrieves N lines from specified opened file.
    """
    return [x for x in islice(file, N)]


def collect_period_ids(lines):
    """
    Collects and groups period ID's from specified lines.
    """
    collected_period_ids = defaultdict(list)

    for line in lines:
        # current period ID  from second field
        field1_end = line.find('|') +1
        field2_end = line.find('|',field1_end)
        period_id = line[field1_end:field2_end]
        # appending current line to list of collected lines associated with
        # current period_id
        collected_period_ids[period_id].append(line)

    return collected_period_ids


def export_grouped_periods(periods):
    """
    Exports collected and grouped periods.
    """
    for period_id in periods:
# defining output file name according to period ID
        out_name = R"D:\tmp\iholding\period%s.txt" % period_id
        # opening output file and writing collected lines to it
        with open(out_name, 'a+') as f:
            f.write("".join(periods[period_id]))


def main():
    
 with open(R"issueholding.txt") as period_src:

    chunk_cnt = 0

    while True:
        # retrieving 1000 input lines at a time
        line_chunk = get_chunk_of_lines(period_src, 1000000)

        # exiting while loop if no more chunk is left
        if not line_chunk:
            break

        chunk_cnt += 1
        print("+ Working on chunk %d" % chunk_cnt)
 
        # collecting, grouping and exporting period ID's
        periods = collect_period_ids(line_chunk)
        export_grouped_periods(periods)
        

# https://ipython.readthedocs.io/en/stable/interactive/magics.html
%prun main()
```
