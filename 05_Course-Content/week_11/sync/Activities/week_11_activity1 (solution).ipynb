```python
import numpy as np
print (np.__version__)
```

```python
np.random.seed(100)
```

```python
#1. Create a dataset of 50 random integers ranging from 0 and 100.

a = np.random.randint(0,100,50)
a
```

```python
#2. Reshape the data to be 10x5.

a.reshape(10,5)
```

```python
a
```

```python
a = a.reshape(10,5)
a
```

```python
#3. Create a dataset of 50 numbers between 0 and 100 using linspace()

b = np.linspace(0,100,50)
b
```

```python
#4. Reshape the data to be 10x5.

b = b.reshape(10,5)
b
```

```python
#print(a)
#print(b)

print(a.shape)
print(b.shape)

print()
vertstk = np.vstack((a,b))
print(vertstk.shape)

hozstk = np.hstack((a,b))
print(hozstk.shape)
```

```python
#5. Use vstack to stack the data for the two arrays call it "vertstk"

vertstk = np.vstack((a,b))
print(vertstk)
print(vertstk.shape)
```

```python
#6. Use hstack to stack the data for the two arrays call it "hozstk"

hozstk = np.hstack((a,b))
print(hozstk)
print(hozstk.shape)
```

```python
#7. for each array divide each element by 2:<br>
#    a. Use a functional tools (such as map or filter) for hozstk then a vectorized approach for vertstk
#    b. for an extra challenge replace all values greater than 25 with 25

## Functional: Map (Note, this is not best practice but can be done.)
print(list(map(lambda x: x/2, hozstk)))
```

```python
## Vectorize:
print(vertstk)
print(vertstk / 2)
```

```python
# b. Here are two methods to cap your data at "25"
arr = vertstk/2
arr[arr > 25] = 25
print(arr)

# Alternatively, A quick "google" can help you find the 'clip' function, which caps maximum values.
print((vertstk / 2).clip(0, 25))
```

```python
#8. get column and row means for hozstk and vertstk
print(vertstk.shape)
print("Col Means")
print(np.mean(vertstk, axis=0))
print("Row Means")
print(np.mean(vertstk, axis=1))
print(type(np.mean(vertstk, axis=1)))
```

```python
print("Col Means")
print(np.mean(hozstk, axis=0))
print("Row Means")
print(np.mean(hozstk, axis=1))
```
