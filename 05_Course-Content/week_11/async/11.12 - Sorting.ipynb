# Sorting a DataFrame

It's worth a minute to talk about sorting DataFrames.  Let's load our gradebook DataFrame.

```python
gradebook = pd.read_csv('gradebook_v2.csv', index_col=0)
gradebook
```

There are two methods to keep in mind.  First, `sort_index()`.

```python
gradebook.sort_index()
```

Notice that this method returns a new sorted DataFrame.  To keep the changes, we could overwrite our `gradebook` variable.

The other main sorting method is `sort_values()`.  To use this method, you need to specify a column to sort over.

```python
gradebook.sort_values('midterm')
```

Notice that missing values get moved to the end.  We may want to reverse the order to put the highest scores at the top.

```python
gradebook.sort_values('midterm', ascending= False)
```

We can also pass in a list to sort by multiple variables.

```python
gradebook.sort_values(['final','midterm'], ascending = False)
```
