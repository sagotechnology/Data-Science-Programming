<p><strong>Spacing versus tabs</strong><br />Use spaces, not tabs.</p>

<p><strong>Consistency in naming</strong>
<ul>
    <li>
variables, functions, ethods, packages, modules use <pre>lower_case_with_underscores</pre>
    </li>
    <li>Classes and Exceptions <pre>CapWords</pre>
    </li>
    <li>Protected methods &amp; internal functions <pre>_single_leading_underscore(self, ...)</pre></li>
    <li>Private methods <pre>__double_leading(self, ...)</pre></li>
    <li>Constants <pre>ALL_CAPS</pre></li>
</ul>

<i>Better:</i>
    <pre>import audio<br />
        core = audio.Core()
        controller = audio.Controller()
    </pre>

<hr />
<i>Consistent naming:</i>
<pre>elements = ...
    elements_active = ...
    elements_defunct = ... </pre>

<p><strong>Importing</strong> ...  The first group is considered better, except for some 3rd party libraries that require/suggest it.
<br />Put imports at the top of the code:
<ol><li>System imports</li>
    <li>Third-party imports</li>
    <li>Local source tree imports</li>
    </ol> to clarify the source of each module.
    </p>
<pre>import gym
import gym.sessions
from gym import sessions</pre>

<i>Avoid:</i>
<pre>from gym import gym_trainer  # symbol from gym/__init__.py
from gym.sessions import gym_session  # symbol from gym/sessions.py
</pre>

 </ul>

<strong>Testing</strong>
<i>General guidelines</i><ul>
    <li>Use long, descriptive names.</li>
    <li>Isolate tests - don't interact with a database or network - test your code, 
        use test data.  Debug one and then the other</li>
    <li>Use factories if possible</li>
    <li>Use asserts to avoid forgetting to fix something!  E.g., <code>assert False, "TODO: finish this part"</code>
    <li>Test your units - better to make sure small bits of code work, using test case.</li>
</ul>
Example:
<pre>
import unittest
import factories

class PersonTest(unittest.TestCase):
    def setUp(self):
        self.person = factories.PersonFactory()
       
    def test_has_age_in_dog_years(self):
        self.assertEqual(self.person.dog_years, self.person.age / 7)
</pre>
<strong>Functional Tests</strong> are higher level, closer to how your user interacts with the code.  Typically used in GUI and web apps.
<p><ul><li>Test with <i>use case scenarios</i>: write out stories of the use of your application.</li>
    <li>Use comments to write out your stories, before writing the test code</li>
    <li>Storytelling like this is now very popular in engineering and software dev</li>
   </ul>
