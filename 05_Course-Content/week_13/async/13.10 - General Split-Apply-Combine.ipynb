Create a function that takes a group returns two rows of data.  The first row should correspond a random reviewer that gave the highest score to the hotel.  The second should correspond to a random reviewer that gave the lowest score.  Test your function on the sample group.

```python
def review_min_max(group):
    return group.sort_values(by = 'score').max().iloc[4:5], group.sort_values(by = 'score').min().iloc[4:5]
```

```python
def top_bot_reviewers(group):
    max_score = group.score.max()
    min_score = group.score.min()
    good_review = group[group.score == max_score].sample(1)
    bad_review = group[group.score == min_score].sample(1)
    return pd.concat([good_review, bad_review])
```

```python
review_min_max(sample_group)
```

```python
top_bot_reviewers(sample_group)
```

Finally, use `apply()` to create a table of top and bottom reviews for all hotels.

```python
by_hotel.apply(top_bot_reviewers)
```
